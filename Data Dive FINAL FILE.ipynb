```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
preliminary_df = pd.read_csv('chm_jail_data.csv', na_values = ['nan', 'Unknown', 'Other'])
preliminary_df
```

```python
preliminary_df['superhighlevel'].unique()
```

```python
preliminary_df.sort_values(by=['jacket_number', 'booking_date'], inplace=True)



preliminary_df['repeat_offenses'] = preliminary_df.groupby('jacket_number')['booking_date'].transform('nunique') - 1

preliminary_df['repeat_offenses'].fillna(0, inplace=True)

preliminary_df
```

```python
preliminary_df[preliminary_df.repeat_offenses > 0]
```

```python
crime_mapping = {
    'DUI': 2,
    'Traffic': 0,
    'Public Order': 1,
    'Violent': 5,
    'Drug': 4,
    'Domestic Violence': 6,
    'Property': 3,
    'Sex': 7
}
```

```python
preliminary_df['crime_level'] = preliminary_df['superhighlevel'].map(crime_mapping)
preliminary_df['crime_level']
```

```python
categorical_cols = ['booking_date', 'released_date', 'arrest_agency', 'marital_status', 'employment_status', 'city', 'race', 'sex', 'STATE', 'citizenship', 'occupation', 'school', 'prisoner_type', 'crime_level', 'repeat_offenses']
num_cols = ['jacket_number', 'age_at_arrest', 'age_at_release', 'hours']

cat_dummies = pd.get_dummies(preliminary_df[categorical_cols], drop_first = True)
df = pd.concat([preliminary_df[num_cols], cat_dummies], axis = 1)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)

from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors = 261)
df = pd.DataFrame(imputer.fit_transform(df),columns = df.columns)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer

imputer = IterativeImputer()
preliminary_df['crime_level'] = imputer.fit_transform(preliminary_df[['crime_level']])
preliminary_df['crime_level'].value_counts()
```

```python
employment_mapping = {
    'Unemployed': 0,
    'Retired': 1,
    'Laid Off': 2,
    'Employed - Part Time': 3,
    'Student': 4,
    'Self Employed': 5,
    'Employed - Full Time': 6,
}
```

```python
preliminary_df['employment_level'] = preliminary_df['employment_status'].map(employment_mapping)
preliminary_df['employment_level']
```
